### Reading data set into python

```python
dta=pd.read_excel('Dataset_w6 (1).xlsx')
dta
```

### calcuating count of countries

```python
dta['country'].count
```

### count of countries wrt to region

```python
dta['Region'].value_counts()
```

### Region with highest GDP

```python
dta1=dta[['Region','GDP per capita (current US$)']].groupby('Region').mean()
```

```python
print(dta1)
```

```python
dta1['GDP per capita (current US$)'].idxmax()
```

NorthernAmerica is the region with  highest GDP

### Region with highest international trade

```python
dta2=dta[['Region','International trade: Imports (million US$)']].groupby('Region').mean()
```

```python
print(dta2)
```

```python
dta2['International trade: Imports (million US$)'].idxmax()
```

NorthernAmerica has highest International trade

### Region with consistent GDP

```python
dta4=dta.groupby('Region').var()
```

```python
dta4
```

```python
dta4['GDP per capita (current US$)'].idxmin()
```

WesternAfrica is the country with most consistent GDP
